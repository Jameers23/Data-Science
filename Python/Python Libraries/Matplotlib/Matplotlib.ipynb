**Matplotlib**

Matplotlib is a popular plotting library in Python used for creating high-quality visualizations and graphs. It offers various tools to generate diverse plots, facilitating data analysis, exploration, and presentation.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title("Weather")
plt.plot(days, max_t, label="Max")
plt.plot(days, min_t, label="Min")
plt.plot(days, avg_t, label="Avg")
plt.legend()
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title("Weather")
plt.plot(days, max_t, label="Max")
plt.plot(days, min_t, label="Min")
plt.plot(days, avg_t, label="Avg")
plt.legend(loc='lower left')
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title("Weather")
plt.plot(days, max_t, label="Max")
plt.plot(days, min_t, label="Min")
plt.plot(days, avg_t, label="Avg")
plt.legend(loc='best', shadow = True)
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title("Weather")
plt.plot(days, max_t, label="Max")
plt.plot(days, min_t, label="Min")
plt.plot(days, avg_t, label="Avg")
plt.legend(loc='best', shadow = True, fontsize = 'small')
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title("Weather")
plt.plot(days, max_t, label="Max")
plt.plot(days, min_t, label="Min")
plt.plot(days, avg_t, label="Avg")
plt.legend(loc='best', shadow = True, fontsize = 'small')
plt.grid()
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title('US Tech Stocks')
plt.bar(company, revenue, label='Revenue')
plt.legend(fontsize='small', loc='best')
```

```python
y = np.arange(len(company))
y
```

```python
plt.xticks(y, company)
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title('US Tech Stocks')
plt.bar(y, revenue, label='Revenue')
plt.legend(fontsize='small', loc='best')
```

```python
profit=[40,2,34,12]
```

```python
plt.xticks(y, company)
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title('US Tech Stocks')
plt.bar(y, revenue, label='Revenue')
plt.bar(y,profit, label='Profit')
plt.legend(fontsize='small', loc='best')
```

```python
y-0.2
```

```python
plt.xticks(y, company)
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title('US Tech Stocks')
plt.bar(y-0.2, revenue, label='Revenue')
plt.bar(y+0.2,profit, label='Profit')
plt.legend(fontsize='small', loc='best')
```

```python
plt.xticks(y, company)
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title('US Tech Stocks')
plt.bar(y-0.2, revenue, width=0.4, label='Revenue')
plt.bar(y+0.2,profit, width=0.4, label='Profit')
plt.legend(fontsize='small', loc='best')
```

```python
plt.yticks(y, company)
plt.ylabel('Company')
plt.xlabel('Revenue')
plt.title('US Tech Stocks')
plt.barh(y, revenue, label='Revenue')
plt.barh(y,profit, label='Profit')
plt.legend(fontsize='small', loc='best')
```

```python
plt.yticks(y, company)
plt.ylabel('Company')
plt.xlabel('Revenue')
plt.title('US Tech Stocks')
plt.barh(y-0.2, revenue, label='Revenue', height=0.4)
plt.barh(y+0.2,profit, label='Profit', height=0.4)
plt.legend(fontsize='small', loc='best')
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
```

```python
plt.hist(blood_sugar)
```

```python
"""
  80-100: Normal
  100-125: Pre-diabetic
  80-100: Diabetic
"""
plt.hist(blood_sugar, bins = 3)
```

```python
plt.hist(blood_sugar, bins = 3, rwidth=0.9)
```

```python
plt.hist(blood_sugar, rwidth=0.8)
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='g')
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='g')
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='g', histtype='step')
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'], label=['men','women'])
plt.legend()
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150],
         rwidth=0.95, color=['green','orange'], label=['men','women'],
         orientation = 'horizontal')
plt.legend()
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.pie(exp_vals, labels=exp_labels)
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels)
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels)
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1)
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1, autopct="%0.2f%%")
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1, autopct="%0.2f%%", shadow=True)
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1, autopct="%0.2f%%", shadow=True, explode=[0,0,0,0,0])
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1, autopct="%0.2f%%", explode=[0,0,0.1,0,0])
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1, autopct="%0.2f%%", explode=[0,0,0.1,0.1,0])
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1, autopct="%0.2f%%", explode=[0,0,0,0.1,0.1])
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1, autopct="%0.2f%%", explode=[0,0,0,0.1,0.1], startangle = 50)
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1, autopct="%0.2f%%", explode=[0,0,0,0.1,0.1], startangle = 50)
plt.savefig('pie1.png', bbox_inches='tight')
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1, autopct="%0.2f%%", explode=[0,0,0,0.1,0.1], startangle = 50)
plt.savefig('pie2.png', bbox_inches='tight', pad_inches = 1)
```
